```python
def histogram_totals (num_cols, columns_to_plot, cobenefit= False):

    num_rows = (len(columns_to_plot) + num_cols ) // num_cols # Ceiling division
    fig, axes = plt.subplots(num_rows, num_cols, figsize=(12, 8))
    axes = axes.flatten() # Flatten the axes array for easier iteration
    
    # Loop through columns and plot histograms
    for i, col in enumerate(columns_to_plot):
        if (col == "WIMD 2025 overall quintile"): 
            discrete = True 
        else: 
            discrete = None

        col_name = col.replace("_", " ").capitalize()
        col_label = col_name

        if (cobenefit): 
            col_name = col_name + " cobenefit"
            col_label = "Total Co-benefit (Â£ million)"
        
        
        sns.histplot(
            data=l2data_totals, 
            x=col, 
            kde=True, 
            ax=axes[i]
            ,discrete= discrete
            )
        axes[i].set_title(f'Distribution of {col_name}') # Set title for each subplot
        axes[i].set_xlabel(col_label) # Set x-axis label
        axes[i].set_ylabel('Number of LSOAs') # Set y-axis label

    # remove any unused subplots
    for j in range(i + 1, len(axes)):
        fig.delaxes(axes[j])

    # Adjust layout and display the plots
    plt.tight_layout()
    plt.show()
    
```

```python
columns_to_plot = [
    'WIMD 2025 overall quintile',
    'population', 
    'households'
    ]
histogram_totals(
    num_cols = 3, 
    columns_to_plot = columns_to_plot
    )
```

```python
columns_to_plot = [
    'dampness',
    'excess_cold', 
    'excess_heat'
    ]
histogram_totals(
    num_cols = 3, 
    columns_to_plot = columns_to_plot
    , cobenefit= True
    )
```

```python
# List the columns you want to sum
cobenefit_columns = [
    'sum',
    'air_quality',
    'congestion',
    'dampness',
    'diet_change',
    'excess_cold',
    'excess_heat',
    'hassle_costs',
    'noise',
    'physical_activity',
    'road_repairs',
    'road_safety'
]
  # add/remove as needed

# Sum each column individually
column_sums = df[cobenefit_columns].sum()

# Convert to DataFrame with proper column names
column_sums = column_sums.reset_index()
column_sums.columns = ['benefit_type', 'value']

# Rename 'sum' to 'total'
column_sums['benefit_type'] = column_sums['benefit_type'].replace('sum', 'total')

# Capitalize and replace underscores with spaces
column_sums['benefit_type'] = column_sums['benefit_type'].str.replace('_', ' ').str.title()


print("Sum for each column:\n", column_sums)
```

```python
tot_net_benefits = round(column_sums.loc[column_sums['benefit_type']=='Total', 'value'],2).values[0]
```

```python
column_sums
```

```python
import plotly.express as px

# Filter out the 'total' row
column_sums_filtered = column_sums[column_sums['benefit_type'] != 'Total']

# Create treemap
fig = px.treemap(
    column_sums_filtered,
    path=['benefit_type'],
    values='value',
    title='Distribution of Co-benefits Across Cardiff LSOAs',
    color='value',
    color_continuous_scale='RdYlGn',
    labels={'value': 'Total Co-benefit (Â£ million)'}
)

fig.update_traces(textinfo='label+value')
fig.update_layout(height=600)
fig.show()
```

```python
# Separate positive and negative values
column_sums_filtered['color_category'] = column_sums_filtered['value'].apply(
    lambda x: 'Positive Co-benefits' if x > 0 else 'Negative Costs'
)

column_sums_filtered
```

```python
# Sort by value for better visualization
column_sums_sorted = column_sums_filtered.sort_values('value', ascending=True)
column_sums_sorted
```

```python
positive_data = column_sums_sorted[column_sums_sorted['value'] > 0]
positive_data
```

```python
# Create diverging bar chart
import plotly.graph_objects as go
fig = go.Figure()

# Create custom text labels with consistent formatting
def format_value(val):
    abs_val = abs(val)
    if abs_val < 0.001 and abs_val > 0:
        return f'Â£{abs_val:.5f}M'  # 5 decimals for very small values
    elif abs_val < 0.1:
        return f'Â£{abs_val:.4f}M'  # 4 decimals for small values
    else:
        return f'Â£{abs_val:.2f}M'  # 2 decimals for normal values

positive_data['formatted_text'] = positive_data['value'].apply(format_value)

fig.add_trace(go.Bar(
    y=positive_data['benefit_type'],
    x=positive_data['value'],
    orientation='h',
    name='Co-benefits',
    marker=dict(color='#2ecc71', line=dict(color='#27ae60', width=1)),
    text=positive_data['formatted_text'],
    textposition='outside',
    textfont=dict(color='black', size=11, family='Arial Black'),
    cliponaxis=False,
    hovertemplate='<b>%{y}</b><br>' +
                'Value: Â£%{x:.5f}M<br>' +
                '<extra></extra>'
))
fig
```
