Looking only at the sum of the co-benefits across the years (column `sum`), transform the data from vertical to horizontal (i.e. one column per co-benefit) to facilitate visual representation

```python
l2data_totals = l2data.pivot(
    index = ['LSOA code', 'LSOA name (Eng)', 'WIMD 2025 overall quintile',
             'population', 'households'],
    columns = 'co-benefit_type',
    values = 'sum'
    ).reset_index()

l2data_totals.head()
```

```python
# save it for further processing
l2data_totals.to_csv("../data/l2data_totals.csv")
```

```python
# Define the columns you want to plot
columns_to_plot = ['WIMD 2025 overall quintile', 'population', 'households']

# 2. Set up the matplotlib figure and axes
# Determine grid dimensions (e.g., 2 rows, 2 columns for 4 plots)
num_cols = 2
num_rows = (len(columns_to_plot) + num_cols) // num_cols # Ceiling division

print(num_cols)

fig, axes = plt.subplots(num_rows, num_cols, figsize=(12, 8))
# Flatten the axes array for easier iteration
axes = axes.flatten()
# 3. Loop through columns and plot histograms
for i, col in enumerate(columns_to_plot):
    if (col == "WIMD 2025 overall quintile"): 
        discrete = True 
    else: discrete = None
    
    sns.histplot(
        data=l2data_totals, 
        x=col, 
        kde=True, 
        ax=axes[i]
        ,discrete= discrete
        )
    axes[i].set_title(f'Distribution of {col} by LSOA') # Set title for each subplot
    axes[i].set_xlabel(col) # Set x-axis label
    axes[i].set_ylabel('Number of LSOAs') # Set y-axis label

# 4. Remove any unused subplots
for j in range(i + 1, len(axes)):
    fig.delaxes(axes[j])

# 5. Adjust layout and display the plots
plt.tight_layout()
plt.show()
```

```python
def histogram_totals (num_cols, columns_to_plot, cobenefit= False):

    num_rows = (len(columns_to_plot) + num_cols ) // num_cols # Ceiling division
    fig, axes = plt.subplots(num_rows, num_cols, figsize=(12, 8))
    axes = axes.flatten() # Flatten the axes array for easier iteration
    
    # Loop through columns and plot histograms
    for i, col in enumerate(columns_to_plot):
        if (col == "WIMD 2025 overall quintile"): 
            discrete = True 
        else: 
            discrete = None

        col_name = col.replace("_", " ").capitalize()
        col_label = col_name

        if (cobenefit): 
            col_name = col_name + " cobenefit"
            col_label = "Total Co-benefit (Â£ million)"
        
        
        sns.histplot(
            data=l2data_totals, 
            x=col, 
            kde=True, 
            ax=axes[i]
            ,discrete= discrete
            )
        axes[i].set_title(f'Distribution of {col_name}') # Set title for each subplot
        axes[i].set_xlabel(col_label) # Set x-axis label
        axes[i].set_ylabel('Number of LSOAs') # Set y-axis label

    # remove any unused subplots
    for j in range(i + 1, len(axes)):
        fig.delaxes(axes[j])

    # Adjust layout and display the plots
    plt.tight_layout()
    plt.show()
    
```

```python
columns_to_plot = [
    'WIMD 2025 overall quintile',
    'population', 
    'households'
    ]
histogram_totals(
    num_cols = 3, 
    columns_to_plot = columns_to_plot
    )
```

```python
columns_to_plot = [
    'dampness',
    'excess_cold', 
    'excess_heat'
    ]
histogram_totals(
    num_cols = 3, 
    columns_to_plot = columns_to_plot
    , cobenefit= True
    )
```
